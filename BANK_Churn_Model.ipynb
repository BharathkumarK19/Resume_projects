```python
#data set
df=pd.read_csv("/content/bank-customer-churn-dataset/Bank Customer Churn Prediction.csv")
```

```python
#pre-processing
df=df.drop('customer_id',axis=1)
```

```python
df['gender'].replace({'Male':1,'Female':0},inplace=True)
```

```python
df['country'].value_counts()
```

```python
df=pd.concat([df,pd.get_dummies(df['country'],prefix='country',drop_first='country')],axis=1)
```

```python
df=df.drop('country'
df=pd.concat([df,pd.get_dummies(df['country'],prefix='country',drop_first='country')],axis=1)
,axis=1)
df.head()
```

```python
x=df.drop('churn',axis=1)
y=df['churn']
```

```python
#splitting the data for training and testing
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=4519)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
#Hyper parameter tunning using GridSearchCV
from sklearn.model_selection import GridSearchCV
```

```python
model=RandomForestClassifier()
model1=GradientBoostingClassifier()
```

```python
model1.fit(x_train,y_train)
```

```python
y_pred=model1.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
param_dis={
    'n_estimators': [50, 100, 150],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]

}
```

```python
grid=GridSearchCV(model1,param_dis,cv=5,n_jobs=-1)
```

```python
grid.fit(x_train,y_train)
```

```python
#Accuracy of model
grid.best_score_
```
